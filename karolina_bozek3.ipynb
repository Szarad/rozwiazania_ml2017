```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
from time import time

n_colors = 3
flower = load_sample_image("flower.jpg")
flower = np.array(flower, dtype=np.float64) / 255


w, h, d = original_shape = tuple(flower.shape)
assert d == 3
image_array = np.reshape(flower, (w * h, d))
t0 = time()
image_array_sample = shuffle(image_array, random_state=0)[:1000]
kmeans = KMeans(n_clusters=n_colors, random_state=0).fit(image_array_sample)

t0 = time()
labels = kmeans.predict(image_array)



codebook_random = shuffle(image_array, random_state=0)[:n_colors + 1]

t0 = time()
labels_random = pairwise_distances_argmin(codebook_random,
                                          image_array,
                                          axis=0)
print("done in %0.3fs." % (time() - t0))


def recreate_image(codebook, labels, w, h):
    d = codebook.shape[1]
    image = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    return image

plt.figure(1)
plt.clf()
ax = plt.axes([0, 0, 1, 1])


plt.imshow(flower)

plt.figure(2)
plt.clf()
ax = plt.axes([0, 0, 1, 1])

plt.imshow(recreate_image(kmeans.cluster_centers_, labels, w, h))
```

```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()


import matplotlib.pyplot as plt
def print_faces(images, target, top_n):
    # set up the figure size in inches
    fig = plt.figure(figsize=(12, 12))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    for i in range(top_n):
        # plot the images in a matrix of 20x20
        p = fig.add_subplot(20, 20, i + 1, xticks=[], yticks=[])
        p.imshow(images[i], cmap=plt.cm.bone)

        # label the image with the target value
        p.text(0, 14, str(target[i]))
        p.text(0, 60, str(i))
    plt.show()
print_faces(faces.images, faces.target, 20)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_similarity_score
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
Xf = faces.data
Yf = faces.target

for i in range(1,7):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(Xf)
    y_true = kmeans.labels_
    y_pred = Yf
    j=jaccard_similarity_score(y_true, y_pred)
    print "jaccard similiarity score", j, "for k=", i
```

```python
import numpy as np
import matplotlib.pylab as plt

def gaussian_2d(x, y, x0, y0, xsig, ysig):
    return np.exp(-0.5*(((x-x0) / xsig)**2 + ((y-y0) / ysig)**2))

def area(vs):
    a = 0
    x0,y0 = vs[0]
    for [x1,y1] in vs[1:]:
        dx = x1-x0
        dy = y1-y0
        a += 0.5*(y0*dx - x0*dy)
        x0 = x1
        y0 = y1
    return a

delta = 0.01
x = np.arange(-3.1, 3.1, delta)
y = np.arange(-3.1, 3.1, delta)
X, Y = np.meshgrid(x, y)
r = np.sqrt(X**2 + Y**2)
levels = [3.0]
cs = plt.contour(X,Y,r,levels=levels)
plt.clabel(cs, inline=1, fontsize=10)

for i in range(len(levels)):
    contour = cs.collections[i]
    vs = contour.get_paths()[0].vertices
    a = area(vs)
    print "r = " + str(levels[i]) + ": Pole powierzchni=" + str(a)

plt.show()
```
